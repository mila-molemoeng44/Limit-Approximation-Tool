```python
import numpy as np
import matplotlib.pyplot as plt

multiplicationFactor = 2
x = np.linspace(-10,10,1000)
y = x** multiplicationFactor
plt.figure(figsize=(8,6))
plt.plot(x,y,label = 'y=x^2')
plt.title("Parabola for my limit")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.axhline(y=0, color='red', linestyle='-') 
plt.axvline(x=0, color='red', linestyle='-')
plt.legend()
plt.show()

x_coord = int(round(x[0]))
x_coord1 = int(round(x[250]))
x_coord2 = int(round(x[500]))
x_coord3 = int(round(x[750]))
x_coord4 = int(round(x[999]))


print(x_coord)
print(x_coord1)
print(x_coord2)
print(x_coord3)
print(x_coord4)
```

```python
#initialising multiplication and rounding numbers
multiplicationFactor = 2
roundingNumber =8

#Asking the user for an limit-corresponding index
answer = int(input("Please provide a limit index for the parabola that exists between 0 and 999"))

#Initialising the x and y values from the graph
x_coord5 = round(x[answer],roundingNumber)
y_coord1 = round(x_coord5**multiplicationFactor, roundingNumber)

#Printing the left and right hand behaviours of the graph near the limit
print("The beahviour of the graph near point x is:")
for i in range(2,-1,-1):
    i = round(x[answer + i],roundingNumber)
    print(i)
    k = round(i**multiplicationFactor,roundingNumber)
    print(k, "\n")
    
    
for j in range (3):
    j = round(x[answer - (j)], roundingNumber)
    print(j)
    g = round(j**multiplicationFactor,roundingNumber)
    print(g, "\n")
        
#Printing the limit and index
print("The limit as the graph approaches ", x_coord5, "corresponding to", answer ," is:",y_coord1)
```
